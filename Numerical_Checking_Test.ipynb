```python
def Numerical_Convert(df):
    Node_List= df["NODE_ID"]
#     for Nodes in Node_List:
#         Node = df[(df["NODE_ID"] == Nodes)]
    for col in df.columns:
        try:
            float(df[col].iloc[0])
            df[col] = pd.to_numeric(df[col], errors = "coerce") 
            df[col] = df[col].replace(-9999.0, np.nan)
            print("Done with column " + str(col) + "...")
        except:
            print("Skipping column " + str(col) + "...")
            continue 
    print("Numerical Conversion Complete!!!")


            
```

```python
def min_max(df):
    for col in df.columns:
        try:
            float(df[col].iloc[0])
            print(str(col) + " Minimum: " + str(df[col].min()) + "\n        Maximum: " + str(df[col].max()) + "\n")
        except:
            print("Skipping column " + str(col) + "...\n")
    print("Minimum Maximum Checking Complete!!!")
```

```python
def col_check(df):
    n = 0
    for col in df.columns:
        n = n + 1
        if "ApogeeO2" in col:
            print(n, "ApogeeO2: " + col)
            
        elif "TEMPERATURE"  in col:
            print(n, "TEMPERATURE: " + col)
            
        elif "DS18"  in col:
            print(n, "DS18: " + col)
            
        elif "HUMIDITY" in col:
            print(n, "HUMIDITY: " + col)
            
        elif "LIGHT" in col:
                print(n, "LIGHT: " + col)
                
        elif "TIME" in col:
            print(n, "TIME: " + col)
            
        elif "SoilMoisture" in col:
            print(n, "SoilMoisture: " + col)
            
        elif "CURRENT" in col:
            print(n, "CURRENT: " + col)
            
        elif "VOLTAGE" in col:
            print(n, "VOLTAGE: " + col)
            
        elif "CO2" in col:
            print(n, "CO2: " + col)
        else:
            print(n, "Skipping column " + col + "...")
    print("\n Column Check Complete!!!")
```

```python
col_check(df1)

print("\n", len(df1.columns))
```
