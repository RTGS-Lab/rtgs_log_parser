```python
import os
from datetime import datetime, timezone
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Merged Dataframes That have Headers (Change to path of dataframe)
df0 = pd.read_csv("Version/1_MergedCSV_0.csv")
df1 = pd.read_csv("Version/1_MergedCSV_1.csv")
df2 = pd.read_csv("Version/1_MergedCSV_2.csv")
df3 = pd.read_csv("Version/1_MergedCSV_3.csv")
df4 = pd.read_csv("Version/1_MergedCSV_4.csv")
```

```python
dfs = [df0, df1, df2, df3, df4]
```

```python
for df in dfs:
    print("\n")
    node_list = df["NODE_ID"].unique()
    for Nodes in node_list:
        Node = df[(df["NODE_ID"] == Nodes)]
        Start = Node["ISO8601_DATE"].iloc[0]
        End = Node["ISO8601_DATE"].iloc[-1]
        print(Nodes + " Start Ping is " + str(Start) + " and End Ping is " + str(End) + "...")
```

```python
for df in dfs:
    print("\n")
    Folder = str(df["MERGE_FOLD_INDX"].iloc[0])
    Time_col = df.columns[6]                 ##Timestamp column
    node_list = df["NODE_ID"].unique()
    for Nodes in node_list:
        Node = df[(df["NODE_ID"] == Nodes)]
        
        Time1 = Node[(Node["ISO8601_DATE"] >= datetime(2019, 3, 16).isoformat())&(Node["ISO8601_DATE"] <= datetime(2020, 3, 15).isoformat())]
        Start = Time1["ISO8601_DATE"].min()
        End = Time1["ISO8601_DATE"].max()
        result = Time1[Time_col].max()-Time1[Time_col].min()
        if not np.isnan(result):
            print(Time1[Time_col].max()-Time1[Time_col].min())
            print(Nodes + " Start Ping is " + str(Start) + " and End Ping is " + str(End) + "...")
        else:
            print(Folder, Nodes, " Has no dates in 2019-2020")
            
            
        
        Time2 = Node[(Node["ISO8601_DATE"] >= datetime(2020, 3, 16).isoformat())&(Node["ISO8601_DATE"] <= datetime(2021, 3, 15).isoformat())]
        Start = Time2["ISO8601_DATE"].min()
        End = Time2["ISO8601_DATE"].max()
        result = Time2[Time_col].max() - Time2[Time_col].min() 
        if not np.isnan(result):
            print(result)
            print(Nodes + " Start Ping is " + str(Start) + " and End Ping is " + str(End) + "...")
        else:
            print(Folder, Nodes, " Has no dates in 2020-2021")
            
    
        Time3 = Node[(Node["ISO8601_DATE"] >= datetime(2021, 3, 16).isoformat())&(Node["ISO8601_DATE"] <= datetime(2022, 3, 15).isoformat())]
        Start = Time3["ISO8601_DATE"].min()
        End = Time3["ISO8601_DATE"].max()
        result = Time3[Time_col].max()-Time3[Time_col].min()
        if not np.isnan(result):
            print(result)
            print(Nodes + " Start Ping is " + str(Start) + " and End Ping is " + str(End) + "...")
        else:
            print(Folder, Nodes, " Has no dates in 2021-2022")
```
