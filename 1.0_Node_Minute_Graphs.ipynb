# Node Delta Minute Graphs

This notebook creates graphs showing diffrences in minutes for each daframe-node package. In this case, a dataframe-node is defined as a unique node with a unique number (or name) of headers. First, an output directory is created for the graphs. Then, a few columns are renamed/added (May not be necessary later). Finally, Seaborn is used to create, name and save the graphs.

```python
import os
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
graph_out = "./Minute_Graphs/"  ## Output of Graphs. (Might want to change initial path.)

if not os.path.exists(graph_out):
    os.mkdir(graph_out)
```

```python
df0 = pd.read_csv("Merged_Data_8_9_2022/concatCSV_0.csv") ## Merged Dataframes (Change to path of dataframe)
df1 = pd.read_csv("Merged_Data_8_9_2022/concatCSV_1.csv")
df2 = pd.read_csv("Merged_Data_8_9_2022/concatCSV_2.csv")
df3 = pd.read_csv("Merged_Data_8_9_2022/concatCSV_3.csv")
df4 = pd.read_csv("Merged_Data_8_9_2022/concatCSV_4.csv")
```

```python
dfs = [df0, df1, df2, df3, df4] ## List of Dataframes
```

## Renameing and Adding Headers. 

This next cell does three tasks. First, it renames the fourth column in each dataframe to "Timestamp" (which represents seconds of package in each dataframe regardless of what it is called). Next, it creates a "Date" column by using the "from timestamp" function from Datetime. Finally, it creates an Index Column which is usful if you are doing any sort function. 

A note that this cell might be unnecessary later. 

```python
for df in dfs:
    df.rename(columns = {df.columns[3]:"TIMESTAMP"}, inplace = True) ##Renames column 3 to "Timestamp"
    Dates = []
    for date in df["TIMESTAMP"]:
 
        Dates.append(datetime.fromtimestamp(date))
    df.insert(3, "DATE", Dates)  #Inserts the date in Column 3 (Timestamp is now column 4)
    df.insert(0, "MERGE_INDEX", df.index) # Perserves the Origional Index... Need This if doing any sort function!
    
    
    
   
```

## Creating Graphs

The final cell creates a delta minutes vs node index graphs. First, a csv variable unique to the directory is created to attach to each title and output file. Next, a node list is created for each unique node in a dataframe and used to split the dataframe into indiviual nodes. Then, A time delta column is created by using the difference of the timestamp column and dividing it by 60 (seconds to minutes). 

A note that the Node index is the same as the Merge index which might need to change in the future.

```python

csv = 0 ## Attached to filename
for df in dfs:
    Node_List = df["NODE_ID"].unique()
    for Nodes in Node_List:
        Node = df[(df["NODE_ID"] == Nodes)] ## Splits df by 
        Node["NODE_INDEX"] = Node.index        ## This still ends up being the Merged index (Change Later)
        Node["TIME_DELTA"] = Node["TIMESTAMP"].diff()/60   ## Creates delta minutes
        ax= sns.scatterplot(x = "NODE_INDEX", y = "TIME_DELTA", data = Node)
        ax.set(xlabel = "Index",
          ylabel = "\u0394 Minutes",                              ## Delta Minutes
           title = "ConcatCSV_" + str(csv) + " " + str(Nodes))  ## Filename Node
        plt.savefig(graph_out + "concatCSV_" + str(csv) + "_" + str(Nodes), bbox_inches = "tight", dpi = 300) ## Filename_Node.png 
        plt.show()
    csv = csv + 1
```
