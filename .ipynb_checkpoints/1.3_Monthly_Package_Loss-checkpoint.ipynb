```python

def Exception_Check(Year, Folder, Device, Exception_col = []):
    print("\n")
    
    Problem_device = Exception

    if Year == 2:
        if Device == "B-9":
            print("Skpping Device " + Device + "...")
        elif Season2["FIELD_ID"].eq(Device).any():
            if Problem_device[Problem_device.eq(Device)].any(1).any():
                Exception_col.append("Field ID " + str(Device) + " has been redeployed!")
                print("Done with " + Folder, Device, "Exception 2021-2022")

            else:
                Exception_col.append(np.nan)
                print("Done with " + Folder, Device, "Exception 2021-2022")
        else:
            print(Folder, "Field ID "  + str(Device) + " was not done in Season 2021-2022...")
    elif Year == 1:
        if Season1["FIELD_ID"].eq(Device).any():
            Exception_col.append(str(Season1.loc[Season1["FIELD_ID"] == Device, "notes"].values[0]))
            print("Done with " + Folder, Device, "Exception 2020-2021")
        else:
            print(Folder, "Field ID "  + str(Device) + " was not done in Season 2020-2021...")


def Ping_Finder(Year, Folder, Nodes, Device, Folder_col = [], Device_col = [], Node_col = [], Startdate = [],  
                Enddate = [], Expected_col = [], Actual_col = [], Rate_col = []):
    if Year == 1:
        try:
            StartDoc = datetime.strptime(Season1.loc[Season1["FIELD_ID"] == Device, "Install Date"].values[0], '%m/%d/%y')
            try:
                EndDoc = datetime.strptime(Season1.loc[Season1["FIELD_ID"] == Device, "Removal Date"].values[0], '%m/%d/%y')
            except:
                EndDoc = pd.to_datetime(Season1["Removal Date"]).max()
            Time1 = Node[(Node["ISO8601_DATE"] >= (datetime(StartDoc.year, StartDoc.month, StartDoc.day, 
                                                            tzinfo = timezone.utc) - timedelta(days = 14)).isoformat()) 
                         & (Node["ISO8601_DATE"] <= (datetime(EndDoc.year, EndDoc.month, EndDoc.day, 
                                                              tzinfo = timezone.utc) + timedelta(days = 14)).isoformat())]
            Start = Time1["ISO8601_DATE"].min()
            End = Time1["ISO8601_DATE"].max()
            result = Time1["TIMESTAMP"].max()-Time1["TIMESTAMP"].min()
            if Folder == str(1):
                print(Folder, Device, result)
            if result == 0:
                print(Folder, Device, "Has only one collected date in in 2020-2021...")
                expected = 0
                actual = len(Time1)                                           
                rate = np.nan                      
                Folder_col.append(Folder)
                Device_col.append(Device)
                Node_col.append(Nodes)
                Startdate.append(Start)
                Enddate.append(End)
                Expected_col.append(expected)
                Actual_col.append(actual)
                Rate_col.append(rate)
            elif not np.isnan(result) and result != 0:
                expected = int(result/(60*15))                                ## Seconds to 15 minute intervals
                actual = len(Time1)                                           ## Length of time dataframe
                rate = round((actual/expected * 100), 2)                      ## Percent  Rate
                Folder_col.append(Folder)
                Device_col.append(Device)
                Node_col.append(Nodes)
                Startdate.append(Start)
                Enddate.append(End)
                Expected_col.append(expected)
                Actual_col.append(actual)
                Rate_col.append(rate)
                print("Done with ", Folder, Device, "2020-2021 dates..." )
            else:
                print(Folder, Device, " Has no collected dates in 2020-2021...")
        except:
            print(Folder, Device, " Has no collected dates in 2020-2021...")
            
    if Year == 2:
        try:
            StartDoc = datetime(2021, 10, 13, tzinfo = timezone.utc)
            EndDoc = datetime.strptime(Season2.loc[Season2["FIELD_ID"] == Device, "Label sent"].values[0], '%m/%d/%y')
            Time2 = Node[(Node["ISO8601_DATE"] >= datetime(StartDoc.year, StartDoc.month, StartDoc.day, 
                                                           tzinfo = timezone.utc).isoformat())
                         & (Node["ISO8601_DATE"] <= (datetime(EndDoc.year, EndDoc.month, EndDoc.day, 
                                                              tzinfo = timezone.utc) + timedelta(days = 14)).isoformat())]
            Start = Time2["ISO8601_DATE"].min()
            End = Time2["ISO8601_DATE"].max()
            result = Time2["TIMESTAMP"].max() - Time2["TIMESTAMP"].min()
            if Folder == str(1):
                print(Folder, Device, result)
            if result == 0:
                print(Folder, Device, "Has only one collected date in in 2021-2022...")
                expected = 0
                actual = len(Time2)                                           
                rate = np.nan                      
                Folder_col.append(Folder)
                Device_col.append(Device)
                Node_col.append(Nodes)
                Startdate.append(Start)
                Enddate.append(End)
                Expected_col.append(expected)
                Actual_col.append(actual)
                Rate_col.append(rate)
            elif not np.isnan(result) and result !=0:
                expected = int(result/(60*15))                                ## Seconds to 15 minute intervals
                actual = len(Time2)                                           ## Length of time dataframe
                rate = round((actual/expected * 100), 2)                      ## Percent  Rate
                Folder_col.append(Folder)
                Device_col.append(Device)
                Node_col.append(Nodes)
                Startdate.append(Start)
                Enddate.append(End)
                Expected_col.append(expected)
                Actual_col.append(actual)
                Rate_col.append(rate)
                print("Done with ", Folder, Device, "2021-2022 dates..." )
            else:
                print(Folder, Device, " Has no collected dates in 2021-2022...")
        except:
            print(Folder, Device, " Has no collected dates in 2021-2022...")
               
    
    

    
            
        
    
```

```python
Exception_col = []
Folder_col = []
Device_col = []
Node_col = []
Startdate = []
Enddate = []
Expected_col = []
Actual_col = []
Rate_col = []

n = 0
for df in dfs:
    Folder = str(df["MERGE_FOLD_INDX"].iloc[0])
    node_list = df["NODE_ID"].unique()
    for Nodes in node_list:
        n = n + 1
        Node = df[(df["NODE_ID"] == Nodes)]
        Device = Node["FIELD_ID"].iloc[0]
        Exception_Check(Year = 1, Folder = Folder, Device = Device, Exception_col = Exception_col)
        Ping_Finder(Year = 1, Folder = Folder, Nodes = Nodes, Device = Device, Folder_col = Folder_col, 
                    Device_col = Device_col, Node_col = Node_col, 
                    Startdate = Startdate,  Enddate = Enddate, 
                    Expected_col = Expected_col, Actual_col = Actual_col, Rate_col = Rate_col)
        Exception_Check(Year = 2, Folder = Folder, Device = Device, Exception_col = Exception_col)
        Ping_Finder(Year = 2, Folder = Folder, Nodes = Nodes, Device = Device, Folder_col = Folder_col, 
                    Device_col = Device_col, Node_col = Node_col, 
                    Startdate = Startdate,  Enddate = Enddate, 
                    Expected_col = Expected_col, Actual_col = Actual_col, Rate_col = Rate_col)
print("\n", len(Exception_col), len(Folder_col), len(Device_col), len(Node_col), len(Startdate), len(Enddate), 
      len(Expected_col), len(Actual_col), len(Rate_col), n)
```
